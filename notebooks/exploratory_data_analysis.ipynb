This notebook explores the Breast Histopathology dataset (IDC patches).

```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import cv2

base_path = '../data/breast_histopathology'
all_images = glob.glob(os.path.join(base_path, '*', '*', '*.png'))
print(fTotal images found: {len(all_images)})
```

```python
def load_sample_images(df, n=5):
    plt.figure(figsize=(15, 5))
    for i, row in df.sample(n).reset_index().iterrows():
        img = cv2.imread(row['path'])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.subplot(1, n, i+1)
        plt.imshow(img)
        plt.title(f"Label: {row['label']}")
        plt.axis('off')
    plt.show()

# Create data list
data = []
for path in all_images[:5000]:
    label = os.path.basename(os.path.dirname(path))
    data.append({'path': path, 'label': label})
df = pd.DataFrame(data)

print("Sample images from dataset:")
load_sample_images(df)
```
